```python
import chess
import random
N = 8
r = []
c = []
for i in range(1, N+1):
    r.append(random.choice(range(1, N+1)))
    c.append(random.choice(range(1, N+1)))
    
    #print('Q{}: {},{}'.format(i,r[-1],c[-1]))
print(r)
print(c)
```

```python
#check repetition of position
counter = 0
for i in range(0,N):
    for j in range(i+1,N):
        if(r[i] == r[j]):
            if(c[i] == c[j]):
                    counter+=1
if counter!=0:
    print("Not Valid. Repetation Found")
else:
    print("Valid. No Repetation")
```
