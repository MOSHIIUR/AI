# BFS

```python
from queue import Queue 

def bfs(startingNode, destinationNode):
    visited = {} # to store expanded nodes //dictionary
    traversal_output = [] # to store the order of traversal //lists
    frontier = Queue()

    #assigning false to each node that they are yet to visited
    # romanian_map.keys(): ---> this command only gives you the citys name. [keys == city]
    for city in romanian_map.keys():
        visited[city] = False        # visited ["Arad"] = False
        
    
    # creating a dictionary, where we are storing information of given Starting_node.
    node = {'City':startingNode,'Distance':0, 'Path':[startingNode]} #dictionry names 'node'
    
    #insert that node in the queue
    #en-Queu 
    frontier.put(node)    # arad.
    
    while not frontier.empty():   #if queue is not empty
        
        # using frontier.get() we are selecting the nodes, then delete it
        #de-Queue
        node = frontier.get()
        print(node)
        
        #finding the city name of the nodes we just got from frontier.get() and storing the name to u
        u = node['City']         
        
        #after finding the name we are making sure that this node has visited
        visited[u] = True
        
        #storing the node in the list called traversal_output
        traversal_output.append(u)
        
        for v in romanian_map[u].keys():
            
            if not visited[v]: # process if v is not expanded                
                
                # talking one keys from all the keys of respective node, and store their distance between them
                d = node['Distance'] + romanian_map[u][v] 
                
                # storing the expanded nodes as a list in p, talking the starting node and adding the nodes operating in loop
                p =  node['Path'] + [v]   
                
                frontier.put({'City':v,'Distance':d,'Path': p})# store the new child node in the frontier
                
                if v == destinationNode: # goal test
                    print("")
                    print("Dstance from Initial to Target: ")
                    print(d)
                    print("")
                    print("NODES VISITED: ")
                    print(traversal_output)

                    
                   
                    return p # breaking from while
    
    print('not found')   
```

```python
class Stack:
    

    def __init__(self):
        self.list = []

    def push(self, item):
        self.list.append(item)

    def pop(self):
        return self.list.pop()

    def top(self):
        return self.list[-1]

    def is_empty(self):
        return len(self.list) == 0


def depth_first_search(graph, start):
    stack = Stack()
    stack.push(start)
    path = []
    


    while not stack.is_empty():
        
        vertex = stack.pop()
        
        
        if vertex in path:
            continue
        path.append(vertex)
        
        for neighbor in graph[vertex].keys():
            stack.push(neighbor)

    return path
```
